### Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3]
y=[4,5,6]
plt.plot(x,y)
plt.show()
```

```python
plt.plot([23,34,56],[78,56,34])
plt.title("Basic linear graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

```python
x=[1,2,3]
y=[4,5,6]
plt.plot(x,y)
plt.grid()
plt.show()
```

```python
# Non-Linear graph
#v=a*t
t=list(range(1,10))
a=list(range(1,len(t)+1))
v=[a[i]*i for i in range(len(t))]
plt.plot(v,t)
plt.show()
```

```python
import numpy as np
```

```python
t=np.arange(0.1,2.0,0.1)
v=[np.sin(i) for i in t]
plt.plot(v,t)
plt.show()
```
